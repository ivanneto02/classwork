```python
import pandas as pd
df = pd.read_csv("pitt-dataset.csv")
```

```python
df["type_patient"].value_counts().plot.pie(title = "type of patients",autopct='%.2f%%')
```

```python
df["type_report"].value_counts().plot.pie(title = "type of reports",autopct='%.2f%%')
```

```python
df["chief_comp"].value_counts().iloc[:10].plot.bar(title = "Top 10 cheif complaints")
```

The most common chief complaint of all patients is "chest pain".

```python
d = df.loc[df["type_patient"] == "CHEST"]
d["code"].value_counts().iloc[:10].plot.pie(title = "Percentage of main codes given the type of patients is 'CHEST'",autopct='%.2f%%')
```
